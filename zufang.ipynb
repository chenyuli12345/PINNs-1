```python
import matplotlib.pyplot as plt
def dianfei(x, season):
    if season == 'summer':
        if x <= 260:
            return x * 0.662
        elif x <= 600:
            return 260 * 0.662 + (x - 260) * 0.712
        else:
            return 260 * 0.662 + 340 * 0.712 + (x - 600) * 0.962
    elif season == 'winter':
        if x <= 200:
            return x * 0.662
        elif x <= 400:
            return 200 * 0.662 + (x - 200) * 0.712
        else:
            return 200 * 0.662 + 200 * 0.712 + (x - 400) * 0.962
    else:
        raise ValueError("Invalid season. Please choose 'summer' or 'winter'.")

def average_dianfei(x):
    summer_fee = dianfei(x, 'summer')
    winter_fee = dianfei(x, 'winter')
    return (summer_fee + winter_fee) / 2

# 示例使用
x = 718.8 # 用电量
average_fee = average_dianfei(x)
print(f"用电量为 {x} 度时，两个季节的平均电费为: {average_fee:.2f} 元")
```

```python
def calculate_usage_chengshi(days_with_less_ac, daily_usage=18, reduced_usage=8):
    """
    计算用电量
    :param days_with_less_ac: 使用空调较少的天数
    :param daily_usage: 每天的正常用电小时为24
    :param reduced_usage: 使用空调较少时的用电小时为12
    :return: 总用电量
    """
    total_days = 30  # 假设一个月有30天
    normal_days = total_days - days_with_less_ac
    total_usage = normal_days * 12 * 1.1 + normal_days * (daily_usage-12) *0.935 + days_with_less_ac * reduced_usage * 0.935 + normal_days * daily_usage * (0.2+0.15+0.05) + days_with_less_ac * reduced_usage * (0.2+0.15+0.05) + 2.5 * 30 * 0.4 + 0.35 * 24 * 30
    # total_usage = normal_days * daily_usage * 1.1 + days_with_less_ac * reduced_usage * 0.935
    return total_usage

# 示例使用
days_with_less_ac = 12  # 使用空调较少的天数，相当于来学校的天数
total_usage = calculate_usage_chengshi(days_with_less_ac)
average_fee = average_dianfei(total_usage) + 3500 - 100 - 100 + 146 + 224 
print(f"使用空调较少的天数为 {days_with_less_ac} 天时，城市天地广场总用电量为: {total_usage:.2f} 度，两个季节的平均费用为: {average_fee:.2f} 元")
```

```python
def calculate_usage_shenzhenbei(days_with_less_ac, daily_usage=24, reduced_usage=12):
    """
    计算用电量
    :param days_with_less_ac: 使用空调较少的天数
    :param daily_usage: 每天的正常用电小时为24
    :param reduced_usage: 使用空调较少时的用电小时为12
    :return: 总用电量
    """
    total_days = 30  # 假设一个月有30天
    normal_days = total_days - days_with_less_ac
    total_usage = normal_days * daily_usage * 0.7 + days_with_less_ac * reduced_usage * 0.7 + 2500  + 146 + 60 + days_with_less_ac * 75 * 2 
    return total_usage

# 示例使用
days_with_less_ac = 12  # 使用空调较少的天数，相当于来学校的天数
total_usage = calculate_usage_chengshi(days_with_less_ac)
average_fee = average_dianfei(total_usage)
print(f"使用空调较少的天数为 {days_with_less_ac} 天时，深圳北总用电量为: {total_usage:.2f} 度，两个季节的平均电费为: {average_fee:.2f} 元")
```
