```python
import matplotlib.pyplot as plt
def dianfei(x, season):
    if season == 'summer':
        if x <= 260:
            return x * 0.662
        elif x <= 600:
            return 260 * 0.662 + (x - 260) * 0.712
        else:
            return 260 * 0.662 + 340 * 0.712 + (x - 600) * 0.962
    elif season == 'winter':
        if x <= 200:
            return x * 0.662
        elif x <= 400:
            return 200 * 0.662 + (x - 200) * 0.712
        else:
            return 200 * 0.662 + 400 * 0.712 + (x - 400) * 0.962
    else:
        raise ValueError("Invalid season. Please choose 'summer' or 'winter'.")

def average_dianfei(x):
    summer_fee = dianfei(x, 'summer')
    winter_fee = dianfei(x, 'winter')
    return (summer_fee + winter_fee) / 2

# 示例使用
x = 259.2 # 用电量
average_fee = average_dianfei(x)
print(f"用电量为 {x} 度时，两个季节的平均电费为: {average_fee:.2f} 元")
```

```python
def calculate_usage_chengshi(days_with_less_ac, daily_usage=24, reduced_usage=8):
    """
    计算用电量
    :param days_with_less_ac: 使用空调较少的天数
    :param daily_usage: 每天的正常用电小时为24
    :param reduced_usage: 使用空调较少时的用电小时为12
    :return: 总用电量
    """
    total_days = 30  # 假设一个月有30天
    normal_days = total_days - days_with_less_ac
    total_usage = normal_days * daily_usage * 1.1 + days_with_less_ac * reduced_usage * 0.935 
    return total_usage

# 示例使用
days_with_less_ac = 12  # 使用空调较少的天数，相当于来学校的天数
total_usage = calculate_usage_chengshi(days_with_less_ac)
average_fee = average_dianfei(total_usage) + 3500 - 100 - 100 + 146 + 60 + days_with_less_ac * 36 * 2 - (days_with_less_ac * 36 * 2 - 400 ) * 0.33
print(f"使用空调较少的天数为 {days_with_less_ac} 天时，城市天地广场总用电量为: {total_usage:.2f} 度，两个季节的平均费用为: {average_fee:.2f} 元")
```

```python
def calculate_usage_dongmen(days_with_less_ac, daily_usage=24, reduced_usage=8):
    """
    计算用电量
    :param days_with_less_ac: 使用空调较少的天数
    :param daily_usage: 每天的正常用电小时为24
    :param reduced_usage: 使用空调较少时的用电小时为12
    :return: 总用电量
    """
    total_days = 30  # 假设一个月有30天
    normal_days = total_days - days_with_less_ac
    total_usage = normal_days * daily_usage * 0.36 + days_with_less_ac * reduced_usage * 0.36 
    return total_usage

# 示例使用
days_with_less_ac = 10  # 使用空调较少的天数
total_usage = calculate_usage_dongmen(days_with_less_ac)
average_fee = average_dianfei(total_usage) + 3800 + 190 + days_with_less_ac * 36 * 2 - (days_with_less_ac * 36 * 2 - 400 ) * 0.33
print(f"使用空调较少的天数为 {days_with_less_ac} 天时，东门天下总用电量为: {total_usage:.2f} 度，两个季节的平均费用为: {average_fee:.2f} 元")
```

```python
def calculate_usage_liancheng(days_with_less_ac, daily_usage=24, reduced_usage=8):
    """
    计算用电量
    :param days_with_less_ac: 使用空调较少的天数
    :param daily_usage: 每天的正常用电小时为24
    :param reduced_usage: 使用空调较少时的用电小时为12
    :return: 总用电量
    """
    total_days = 30  # 假设一个月有30天
    normal_days = total_days - days_with_less_ac
    total_usage = normal_days * daily_usage * 0.65 + days_with_less_ac * reduced_usage * 0.65
    return total_usage

# 示例使用
days_with_less_ac = 10  # 使用空调较少的天数
total_usage = calculate_usage_dongmen(days_with_less_ac)
average_fee = average_dianfei(total_usage) + 4300 + 140 + days_with_less_ac * 26 * 2
print(f"使用空调较少的天数为 {days_with_less_ac} 天时，东门天下总用电量为: {total_usage:.2f} 度，两个季节的平均费用为: {average_fee:.2f} 元")
```

```python
# 生成数据
days_range = range(10, 19)  # 从0到30天
average_fees_chengshi = []
average_fees_dongmen = []
average_fees_liancheng = []
fee_differences_cd = []
fee_differences_cl = []
fee_differences_ld = []

for days_with_less_ac in days_range:
    total_usage_chengshi = calculate_usage_chengshi(days_with_less_ac)
    average_fee_chengshi = average_dianfei(total_usage_chengshi) + 3500 - 100 - 100 + 146 + 224 + days_with_less_ac * 36 * 2 - (days_with_less_ac * 36 * 2 - 400 ) * 0.33
    average_fees_chengshi.append(average_fee_chengshi)
    
    total_usage_dongmen = calculate_usage_dongmen(days_with_less_ac)
    average_fee_dongmen = average_dianfei(total_usage_dongmen) + 3800 + 190 + 160 + days_with_less_ac * 36 * 2 - (days_with_less_ac * 36 * 2 - 400 ) * 0.33
    average_fees_dongmen.append(average_fee_dongmen)

    total_usage_liancheng = calculate_usage_liancheng(days_with_less_ac)
    average_fee_liancheng = average_dianfei(total_usage_liancheng) + 4300 + 140 + 180 + days_with_less_ac * 26 * 2
    average_fees_liancheng.append(average_fee_liancheng)

    fee_difference_cd = (average_fee_chengshi - average_fee_dongmen) * 12
    fee_differences_cd.append(fee_difference_cd)

    fee_difference_cl = (average_fee_chengshi - average_fee_liancheng) * 12
    fee_differences_cl.append(fee_difference_cl)

    fee_difference_ld = (average_fee_liancheng - average_fee_dongmen) * 12
    fee_differences_ld.append(fee_difference_ld)


# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(days_range, average_fees_chengshi, marker='o', label='chengshi')
plt.plot(days_range, average_fees_dongmen, marker='x', label='dongmen')
plt.plot(days_range, average_fees_liancheng, marker='s', label='liancheng')
plt.xlabel('laixuexiaotianshu')
plt.ylabel('yuefeiyong')
plt.title('不同使用空调较少的天数下的平均电费')
plt.legend()
plt.grid(True)

# 标注电费差值
for i, days in enumerate(days_range):
    plt.annotate(f'{fee_differences_cd[i]:.2f}', (days, (average_fees_chengshi[i] + average_fees_dongmen[i]) / 2),
                 textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='red')
    plt.annotate(f'{fee_differences_cl[i]:.2f}', (days, (average_fees_chengshi[i] + average_fees_liancheng[i]) / 2),
                    textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='blue')
    plt.annotate(f'{fee_differences_ld[i]:.2f}', (days, (average_fees_liancheng[i] + average_fees_dongmen[i]) / 2),
                textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='green')

plt.show()
```

```python
def calculate_usage_shenzhenbei(days_with_less_ac, daily_usage=24, reduced_usage=12):
    """
    计算用电量
    :param days_with_less_ac: 使用空调较少的天数
    :param daily_usage: 每天的正常用电小时为24
    :param reduced_usage: 使用空调较少时的用电小时为12
    :return: 总用电量
    """
    total_days = 30  # 假设一个月有30天
    normal_days = total_days - days_with_less_ac
    total_usage = normal_days * daily_usage * 0.7 + days_with_less_ac * reduced_usage * 0.7 + 2500  + 146 + 60 + days_with_less_ac * 75 * 2 
    return total_usage

# 示例使用
days_with_less_ac = 12  # 使用空调较少的天数，相当于来学校的天数
total_usage = calculate_usage_chengshi(days_with_less_ac)
average_fee = average_dianfei(total_usage)
print(f"使用空调较少的天数为 {days_with_less_ac} 天时，深圳北总用电量为: {total_usage:.2f} 度，两个季节的平均电费为: {average_fee:.2f} 元")
```

```python
# 生成数据
days_range = range(10, 19)  # 从0到30天
average_fees_chengshi = []
average_fees_dongmen = []
average_fees_shenzhenbei = []
fee_differences = []

for days_with_less_ac in days_range:
    total_usage_chengshi = calculate_usage_chengshi(days_with_less_ac)
    average_fee_chengshi = average_dianfei(total_usage_chengshi)
    average_fees_chengshi.append(average_fee_chengshi)
    
    total_usage_dongmen = calculate_usage_dongmen(days_with_less_ac)
    average_fee_dongmen = average_dianfei(total_usage_dongmen)
    average_fees_dongmen.append(average_fee_dongmen)

    total_usage_shenzhenbei = calculate_usage_shenzhenbei(days_with_less_ac)
    average_fee_shenzhenbei = average_dianfei(total_usage_shenzhenbei)
    average_fees_shenzhenbei.append(average_fee_shenzhenbei)

    fee_difference = (average_fee_chengshi - average_fee_dongmen) * 12
    fee_differences.append(fee_difference)

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(days_range, average_fees_chengshi, marker='o', label='chengshitiandi')
plt.plot(days_range, average_fees_dongmen, marker='x', label='dongmentianxia')
plt.plot(days_range, average_fees_shenzhenbei, marker='s', label='shenzhenbei')
plt.xlabel('laidetianshu')
plt.ylabel('yuehuafei')
# plt.title('不同使用空调较少的天数下的平均电费')
plt.legend()
plt.grid(True)

# 标注电费差值
for i, days in enumerate(days_range):
    plt.annotate(f'{fee_differences[i]:.2f}', (days, (average_fees_chengshi[i] + average_fees_dongmen[i]) / 2),
                 textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='red')

plt.show()
```
